```python
df1.fillna(method='ffill' , inplace=True)
```

```python
df1["text"] = df1["text"].astype(str)
df1["label"] = df1["label"].astype(str)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
sentiment_le = LabelEncoder()
```

```python
y = sentiment_le.fit_transform(df1['label'])
y
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    stop_words="english",   # 🔥 removes stopwords
    max_features=7000,
    ngram_range=(1, 2)
)

x = tfidf.fit_transform(df1["text"])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y  , test_size=0.2 , random_state=42)
```

```python
x_train.shape , x_test.shape , y_train.shape , y_test.shape
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
sent_lr = LogisticRegression()
```

```python
sent_lr.fit(x_train , y_train)
```

```python
y_pred = sent_lr.predict(x_test)
y_pred
```

```python
y_pred_ = sentiment_le.inverse_transform(y_pred)
print(y_pred_[:10])
```

### Model Evaluation

```python
from sklearn.metrics import accuracy_score , f1_score
```

```python
accuracy_score(y_pred , y_test)
```

```python
f1_score(y_pred , y_test , average='micro')
```
